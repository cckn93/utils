```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 2, 100)
density = 2 * x
temperature = 4 - 2 * x
velocity = np.linspace(60, 0, 100)

# Create a figure and axis
fig, ax1 = plt.subplots()

# Plot Density on the primary y-axis
ax1.plot(x, density, 'b-', label='Density')
ax1.set_xlabel('Distance')
ax1.set_ylabel('Density', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Create a secondary y-axis and plot Temperature
ax2 = ax1.twinx()
ax2.plot(x, temperature, 'orange', label='Temperature')
ax2.set_ylabel('Temperature', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Optionally, add another y-axis for Velocity
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))  # Offset the third axis
ax3.plot(x, velocity, 'g-', label='Velocity')
ax3.set_ylabel('Velocity', color='green')
ax3.tick_params(axis='y', labelcolor='green')

# Add a legend
fig.legend(loc='upper left', bbox_to_anchor=(0.15, 0.85))

# Show the plot
plt.show()
```
